```python
# Import libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Filepaths:

h_samples_path = "outputs/h_samples.csv"
sigma_samples_path = "outputs/sigma_samples.csv"
scale_samples_path = "outputs/scale_samples.csv"
```

```python
# Import Data
h_samples = pd.read_csv(h_samples_path)
h_samples

sigma_samples = pd.read_csv(sigma_samples_path)
sigma_samples

scale_samples = pd.read_csv(scale_samples_path)
scale_samples
```

```python
# Print the min, max, and mean values of the h_samples
print("h Min:", h_samples.min(axis=None))
print("h Max:", h_samples.max(axis=None))
print("h Mean:", h_samples.mean(axis=None))
print("\n")

# Print the min, max, and mean values of the sigma_samples
print("sigma Min:", sigma_samples.min(axis=None))
print("sigma Max:", sigma_samples.max(axis=None))
print("sigma Mean:", sigma_samples.mean(axis=None))
print("\n")

# Print the min, max, and mean values of the scale_samples
print("scale Min:", scale_samples.min(axis=None))
print("scale Max:", scale_samples.max(axis=None))
print("scale Mean:", scale_samples.mean(axis=None))
print("\n")
```
