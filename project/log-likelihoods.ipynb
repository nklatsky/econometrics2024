# Analyze Log-Likelihood Under SV Model

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Current working directory: ", os.getcwd())
```

```python
# Read in log_likelihoods for SV Model
sv_log_likelihoods = pd.read_csv("stochastic_volatility/sampler_outputs/log_likelihood_samples.csv")
sv_log_likelihoods
```

```python

def compute_total_log_likelihood(log_likelihoods):
    # Calculate the total log-likelihood
    total_log_likelihood = np.sum(log_likelihoods)

    # Return the total likelihood for further use
    return total_log_likelihood

```

```python
# Define indices to test
sv_indices = [0, 100, 200, 300, 400]

for idx in sv_indices:
    row = sv_log_likelihoods.iloc[idx]
    total_log_likelihood = compute_total_log_likelihood(row)
    print(f"Total log-likelihood for row {idx}: {total_log_likelihood}")
    print("\n")

```

# Analyze Log-Likelihoods Under GARCH(1,1) Model

```python
# Read in log_likelihoods for SV Model
garch_log_likelihoods = pd.read_csv("garch/sampler_outputs/log_likelihood_samples.csv")
garch_log_likelihoods
```

```python
# Define indices to test
garch_indices = [0, 100, 200, 300, 400]

for idx in garch_indices:
    row = garch_log_likelihoods.iloc[idx]
    total_log_likelihood = compute_total_log_likelihood(row)
    print(f"Total log-likelihood for row {idx}: {total_log_likelihood}")
    print("\n")




```

# Comparison


```python
# Compute total_log_likelihoods for all rows, store in a list, and compute the mean and StdDev

garch_rows_log_p = []
for row in garch_log_likelihoods.iterrows():
    total_log_likelihood = compute_total_log_likelihood(row[1])
    garch_rows_log_p.append(total_log_likelihood)


sv_rows_log_p= []
for row in sv_log_likelihoods.iterrows():
    total_log_likelihood = compute_total_log_likelihood(row[1])
    sv_rows_log_p.append(total_log_likelihood)

# garch_rows_log_p
# sv_rows_log_p

garch_avg_log_likelihood = np.mean(garch_rows_log_p)
garch_std_dev_log_likelihood = np.std(garch_rows_log_p)

sv_avg_log_likelihood = np.mean(sv_rows_log_p)
sv_std_dev_log_likelihood = np.std(sv_rows_log_p)

print(f"Average log-likelihood for GARCH model: {garch_avg_log_likelihood}")
print(f"Standard deviation of log-likelihood for GARCH model: {garch_std_dev_log_likelihood}")
print("\n")

print(f"Average log-likelihood for SV model: {sv_avg_log_likelihood}")
print(f"Standard deviation of log-likelihood for SV model: {sv_std_dev_log_likelihood}")
print("\n")


# Create bar chart with error bars
models = ["GARCH(1,1)", "Stochastic Volatility"]
avg_log_likelihoods = [-1 * garch_avg_log_likelihood, -1 * sv_avg_log_likelihood]
std_dev_log_likelihoods = [garch_std_dev_log_likelihood, sv_std_dev_log_likelihood]

fig, ax = plt.subplots()
ax.bar(models, avg_log_likelihoods, yerr=std_dev_log_likelihoods, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Log-likelihood')
ax.set_title('Average negative log-likelihood by model')
ax.yaxis.grid(True)
plt.show()
```
