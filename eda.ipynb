```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get data

data_path = "project/code/fx_data_approx_percent_returns.csv"
data = pd.read_csv(data_path)

# Define dependent variable
y = data["DXY"].values
y

#print diagnostic info about the series y
print("Mean: ", np.mean(y))
print("Standard Deviation: ", np.std(y))
print("Minimum: ", np.min(y))
print("Maximum: ", np.max(y))

print('log std dev: ', np.log(np.std(y)))   
```

```python
# Plot the series
plt.plot(y)
plt.title("DXY returns")
plt.show()
# Plot the absolute value of the series
plt.plot(np.abs(y))
# Plot the mean of the absolute value of the series as a dotted line
plt.axhline(y=np.mean(np.abs(y)), color='r', linestyle='--')
plt.title("Absolute DXY returns")
plt.show()

# Print quartiles of the absolute values of the series
print("25th percentile: ", np.percentile(np.abs(y), 25))
print("50th percentile: ", np.percentile(np.abs(y), 50))
print("75th percentile: ", np.percentile(np.abs(y), 75))
print("90th percentile: ", np.percentile(np.abs(y), 90))
print("95th percentile: ", np.percentile(np.abs(y), 95))
```

```python
# Plot and print percentiles of the differenced absolute value series
plt.plot(np.diff(np.abs(y)))
plt.title("Differenced absolute DXY returns")
plt.show()

# take the absolute value of the differenced series
abs_diff_abs_y = np.abs(np.diff(np.abs(y)))

print("25th percentile: ", np.percentile(abs_diff_abs_y, 25))
print("50th percentile: ", np.percentile(abs_diff_abs_y, 50))
print("75th percentile: ", np.percentile(abs_diff_abs_y, 75))
print("90th percentile: ", np.percentile(abs_diff_abs_y, 90))
print("95th percentile: ", np.percentile(abs_diff_abs_y, 95))
```
